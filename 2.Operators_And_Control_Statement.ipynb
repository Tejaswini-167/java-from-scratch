# Operators And Control Statement

## Arithmetic Operators
#### Arithmetic Operators are used to perform simple arithmetic operations on primitive and non-primitive data types. 

* (*) : Multiplication
* (/) : Division
* (%) : Modulo
* (+) : Addition
* (-) : Subtraction

```java
float x=10;
float y=3;
System.out.println("Sum="+ (x+y));
System.out.println("Sub="+ (x-y));
System.out.println("Div="+ (x/y));
System.out.println("Product="+ (x*y));
System.out.println("Modulo="+ (x%y));
```

## Relational Operators
#### Are used to check some relationship between two operends

Relational operators compare values and return Boolean results:

* (==) : Equal to.
* (!=) : Not equal to.
* (<)  : Less than.
* (<=) : Less than or equal to
* (>)  : Greater than.
* (>=) : Greater than or equal to

```java
   int a = 10;
        int b = 3;
        int c = 5;

        System.out.println("a > b: " + (a > b));
        System.out.println("a < b: " + (a < b));
        System.out.println("a >= b: " + (a >= b));
        System.out.println("a <= b: " + (a <= b));
        System.out.println("a == c: " + (a == c));
        System.out.println("a != c: " + (a != c));
```

## Assingment Operator
####  '=' The assingment operator is used to assign a value to any variable

* += , Add and assign.
* -= , Subtract and assign.
* *= , Multiply and assign.
* /= , Divide and assign.
* %= , Modulo and assign.

```java
 int f = 7;
        System.out.println("f += 3: " + (f += 3));
        System.out.println("f -= 2: " + (f -= 2));
        System.out.println("f *= 4: " + (f *= 4));
        System.out.println("f /= 3: " + (f /= 3));
        System.out.println("f %= 2: " + (f %= 2));
```

## Ternary Operator

#### The Ternary Operator is a shorthand version of the if-else statement. It has three operands and hence the name Ternary. The general format is,

### condition ? if true : if false 

```java
Scanner sc=new Scanner(System.in);
int n=sc.nextInt();
String result;
result=(n%2==0)? "even" :"odd" ;
System.out.println(result);
```

## Unary Operators
####  Unary Operators need only one operand. They are used to increment, decrement, or negate a value. 

* (-): Negates the value
+ (+): Indicates a positive value 
* (++): Increments by 1
    * Post-Increment: Uses value first, then increments
    * Pre-Increment: Increments first, then uses value
* (--): Decrements by 1
    * Post-Decrement: Uses value first, then decrement
    * Pre-Decrement: Decrements first, then uses value
* (!): Inverts a boolean value

```java
 int a = 10;
int b = 10;

// Using unary operators
System.out.println("Postincrement : " + (a++));
System.out.println("Preincrement : " + (++a));

System.out.println("Postdecrement : " + (b--));
System.out.println("Predecrement : " + (--b));
```

## Logical Operators
#### Logical Operators are used to perform "logical AND" and "logical OR" operations, similar to AND gate and OR gate in digital electronics. 

- &&, Logical AND: returns true when both conditions are true.
- ||,  Logical OR: returns true if at least one condition is true.
- !,  Logical NOT: returns true when a condition is false and vice-versa

# Control Statement
### Control statements are used to control the flow of execution of a program based on certain conditions
#### Java provides several control statements to manage program flow, including:

+ Conditional Statements: if, if-else, nested-if, if-else-if
+ Switch-Case: For multiple fixed-value checks
+ Jump Statements: break, continue, return





## If Statement
#### The if statement is the most simple decision-making statement. It is used to decide whether a certain statement or block of statements will be executed or not

```java
Scanner sc=new Scanner(System.in);
int num= sc.nextInt();

if(num > 10)
    System.out.println("number greater than 10");
System.out.println("executed succesfully");

```

### If Else Statements
#### if-else statements are conditional control structures that allow a program to execute different blocks of code based on whether a specified condition is true or false. 

```java
Scanner sc=new Scanner(System.in);
int age=sc.nextInt();
if (age>18)
    System.out.println("Eligible to vote");
else 
    System.out.println("not eligible");
    
    
```

### Nested if Statement
#### A nested if statement in Java is an if statement placed inside another if  block

```java
Scanner sc=new Scanner(System.in);
int num=sc.nextInt();
if (num % 2 ==0) 
    System.out.println("number is an even number");
    if (num%5==0)
        System.out.println("number divisible by 5");
        if (num%10==0)
             System.out.println("number divisible by 10");
```

### if else if Statement
#### An if-else if statement, also known as an if-else if ladder, allows a program to test a series of conditions sequentially

```java
Scanner sc=new Scanner(System.in);
String weather =sc.nextLine();
if (weather.equals("Sunny"))
    System.out.println("carry a sunglass");
else if (weather.equals("rainy"))
    System.out.println("carry an umbrella");
else
    System.out.println("check the weather forecast");
```

 ### Switch Case
 #### The switch statement is a multiway branch statement. It provides an easy way to dispatch execution to different parts of code based on the value of the expression. 

```java
Scanner sc=new Scanner(System.in);
int day =sc.nextInt();
switch(day){
    case 1:
        System.out.println("Sunday");
        break;
            
    case 2:
        System.out.println("Monday");
        break;
    case 3:
        System.out.println("Tuesday");
        break;
    case 4:
        System.out.println("Wednesday");
        break;
    case 5:
        System.out.println("Thursday");
        break;
    case 6:
        System.out.println("Friday");
        break;
    case 7:
        System.out.println("Saturday");
        break;
    default:
        System.out.println("invalid number");
}
            
```

### Jump Statements
#### Jump statements are used to transfer control from one part of the program to anotherThere are three main jump statements in Java
#### 1. break Statement
* Exits from a loop or a switch block immediately.
* Control jumps to the statement after the loop/switch.
  
#### 2. continue Statement
* Skips the current iteration of a loop and jumps to the next iteration.

#### 3. return Statement
* Exits from the current method and returns control to the caller.
* Can also return a value (if the method is not void).
* Doesnt terminate the loop, just skips.
