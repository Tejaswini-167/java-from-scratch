## Typecasting
* Typecasting refers to changing the type of data from one type to another
### types
#### Widening Type casting<small> (Smaller datatype to larger datatype\)<small>
  * Automatically done by java
#### Explicit Type casting<small>(larger datatype to smaller datatype\)<small>
  * User Defined

```java
// widening typecasting
int i=200;
long l=i;
System.out.println(l);
```

```java
// Explicit type casting
long x =500000L;
int y= (int)x;
System.out.println(y);
```

```java
//  final keyword is used to lock the constants so that they can't change
 final float PI = 3.14F;
System.out.println(PI);

PI = 3;   
System.out.println(PI);
```
